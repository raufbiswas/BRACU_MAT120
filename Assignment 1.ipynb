#Question1

1. (7 points) Find the minima and maxima of the following function at a given interval:

```
y = x^4 − 2/3*x^3 − 2x^2 + 2x
in the interval [0,3]
```

Hints: You may want to use conditional statement to gatekeep the values. However, do not
use solveset() function.

````python
import sympy as sp
from sympy import *
sp.init_printing()
````

````python
x = sp.symbols('x')
expr = x**4 - sp.Rational(2,3)*x**3 - 2*x**2 + 2*x
expr
````

````python
dif1 = sp.diff(expr,x)
dif1
````

````python
check = sp.solve(dif1,x)
check
````

````python
#interval [0,3]
list1 = []
for i in check:
  if i>=0 and i<=3:
    list1.append(i)
list1
````

````python
dif2 = sp.diff(expr,x,2)
dif2
````

````python
mm1=dif2.subs(x,list1[0])
mm2=dif2.subs(x,list1[1])

mm1,mm2
````

````python
if mm1<0:
    max = list1[0]
    min = list1[1]
elif mm1>0:
    max = list1[1]
    min = list1[0]
max, min
````

````python
R_max = round(expr.subs(x, max).evalf(),2)
R_mini = round(expr.subs(x, min).evalf(),2)

R_max, R_mini
````

````python
print(f"In the given interval [0,3],\nThe minima and maxima of the following function are respectably {R_mini} and {R_max}.")
````

2.(a) (9 points) Find the minima and maxima of the following functions at a given interval: y = sin(x) in the interval [2π, 4π].

(b) (4 points) Furthermore, determine the x values at which maxima and minima occurs.

Hints: Please refer to the solveset(equation,variable,Interval(num1,num2)) function to obtain the value within the given interval. However, solveset() does not return the values as an array. Therefore, it may require further processing. Additionally, π is accessed in SymPy as “sp.pi”

````python
import sympy as sp
from sympy import *
sp.init_printing()
````

````python
#2_a
x = sp.symbols('x')
expr2a = sp.sin(x)
expr2a
````

````python
dif1 = sp.diff(expr2a,x)
dif1
````

````python
check = solveset(dif1,x, Interval(2*sp.pi, 4*sp.pi))
check = list(check)
check
````

````python
dif2 = sp.diff(expr2a,x,2)
dif2
````

````python
mm1 = dif2.subs(x,check[0])
mm2 = dif2.subs(x,check[1])

mm1,mm2
````

````python
if mm1<0:
    max = check[0]
    min = check[1]
elif mm1>0:
    max = check[1]
    min = check[0]
max, min
````

````python
R_max = expr2a.subs(x, max)
R_mini = expr2a.subs(x, min)

R_max, R_mini
````

````python
print(f"In the given interval [2π, 4π],\nThe minima and maxima of the following function are respectably {R_mini} and {R_max}.")
````
