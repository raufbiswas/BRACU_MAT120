```python
import scipy
from scipy.integrate import odeint
import numpy as np
```

```python
dtheta_dt = lambda t,theta: -2.2067e-12*(theta**4-81e8)
t_val1 = np.linspace(0,480,480)
solve1 = odeint(dtheta_dt, y0= 1200, t= t_val1, tfirst= True)
solve1.shape
```

```python
expr1 = lambda t,v,g,b: g-b*v**2
result = np.zeros((5,100))
t_val2 = np.linspace(0,100,100)
b_1 = np.array([-4,-2,1,2,3])
for i in range(len(b_1)):
    solve2 = odeint(expr1, y0= 0, t = t_val2, args= (10,b_1[i]), tfirst = True)
    result[i,] = solve2.T
result.shape
```
