#Numerical Integration

**SYMPY**

```python
import sympy as sp
from sympy import *
sp.init_printing()
x,y,z = sp.symbols('x y z')
```

```python
expr1 = sp.exp(x)
expr1
```

```python
solve1 = sp.integrate(expr1, (x,0,1))
solve1
```

```python
#sympy can't deal with this equation
expr2 = x**x
expr2
```

```python
solve2 = sp.integrate(expr2, (x,0,1))
solve2
```

```python
#sympy can't deal with this equation
expr3 = x**sp.sin(x)
expr3
```

```python
solve3 = sp.integrate(expr3, (x,0,sp.pi/2))
solve3
```

**SCIPY**

```python
import scipy
from scipy.integrate import quad, fixed_quad
import numpy as np
```

**GAUSSIAN QUADRATURE**

**QUAD(LAMBDIFY)**

```python
#with error value
expr4 = sp.lambdify(x,expr2)
solve4 = quad(expr4,0,1)
solve4
```

```python
#without error
solve4[0]
```

```python
#with error value
expr5 = sp.lambdify(x,expr3)
solve5 = quad(expr5,0,np.pi/2)
solve5
```

```python
#without error
solve5[0]
```

**QUAD(LAMBDA)**

```python
expr6 = lambda x: x**(np.sin(x))
solve6 = quad(expr6,0,np.pi/2)
solve6[0]
```

**FIXED_QUAD(LAMBDA)**

```python
solve7 = fixed_quad(expr6,0,np.pi/2,n=4)
solve7
```

```python
solve7[0]
```

**QUAD(LAMBDA)**

```python
expr7 = lambda x: ( (3*np.exp(x)) + (x**2) + (1/(x+2)) )
solve8 = quad(expr7,-1,1)
solve8
```

```python
solve8[0]
```

**FIXED_QUAD(LAMBDA)**

```python
solve9 = fixed_quad(expr7,-1,1,n=3)
solve9
```

```python
solve9[0]
```
