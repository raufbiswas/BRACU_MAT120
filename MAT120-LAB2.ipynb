```python
import sympy as sp
from sympy import * #to import everything
sp.init_printing()
x,y,z = sp.symbols('x y z')
expr1 = x**3 - 9*x**2 + 8*x+4
#expr1
expr2 = sp.Eq(x**3 - 9*x**2 + 8*x+4,0)
expr2
```

```python
#differentiate
import sympy as sp
from sympy import * #to import everything
sp.init_printing()
x,y,z = sp.symbols('x y z')
expr1 = x**3 - 9*x**2 + 8*x+4
der1 = sp.diff(expr1,x)
#der
der2 = sp.diff(der1,x) #method 1
#der2
der3 = sp.diff(expr1,x,3) #method 2
der3
```

```python
#solve an equation
import sympy as sp
from sympy import * #to import everything
sp.init_printing()
x,y,z = sp.symbols('x y z')
expr1 = x**3 - 9*x**2 + 8*x+4
der1 = sp.diff(expr1,x) #differentiate
sol = sp.solve(der1,x) #solve an equation
sol
```

```python
#max
import sympy as sp
from sympy import * #to import everything
sp.init_printing()
x,y,z = sp.symbols('x y z')
expr1 = x**3 - 9*x**2 + 8*x+4
max = expr1.subs(x,2)
max
```
