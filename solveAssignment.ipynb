```python
#1a
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, trapz
from scipy.misc import derivative
def f(x):
    return 2*x*(x-3)*(np.sin(x)-5)

def h(t):
    return t**2
def dfdx(x):
    return derivative(f, x)
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
t = np.linspace(-4, 4, 1000)

plt.figure(figsize=(12, 6))

# Plot f(x)
plt.subplot(121)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, dfdx(x), label="f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f(x) and its derivative')
plt.legend()

# Plot h(t)
plt.subplot(122)
plt.plot(t, h(t), label='h(t)')
plt.plot(t, derivative(h, t), label="h'(t)")
plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of h(t) and its derivative')
plt.legend()

plt.show()
```

```python
#2a
import numpy as np

def f1(x):
    return 1 / (1 + x**2)

def f2(x):
    return x**2 + 2*x + 1

def simpsons_rule(f, a, b, N):
    h = (b - a) / N
    xi = a
    sum = f(a) + f(b)

    for i in range(1, N):
        xi += h
        if i % 2 == 0:
            sum += 2 * f(xi)
        else:
            sum += 4 * f(xi)

    return (h / 3) * sum

# Define the required parameters
a = 0
b = 4
N = 1000

# Compute the integrals
integral1 = simpsons_rule(f1, a, b, N)
integral2 = simpsons_rule(f2, a, b, N)

print(f"The integral of f1(x) from {a} to {b} using Simpson's Rule is: {integral1}")
print(f"The integral of f2(x) from {a} to {b} using Simpson's Rule is: {integral2}")
```

```python
#2b
from scipy.integrate import quad

def f1(x):
    return 1 / (1 + x**2)

def f2(x):
    return x**2 + 2*x + 1

# Compute the integrals
integral1, err1 = quad(f1, 0, 4)
integral2, err2 = quad(f2, 0, 1)

print(f"The integral of f1(x) from 0 to 4 using scipy.integrate is: {integral1}")
print(f"The error in the approximation is: {err1}")
print(f"The integral of f2(x) from 0 to 1 using scipy.integrate is: {integral2}")
print(f"The error in the approximation is: {err2}")
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt
# Parameters
L = 20
x = np.linspace(-L, L, 1000)
dx = x[1] - x[0]
# Trapezoidal Rule
C = np.trapz(np.cos(x**2), dx=dx)
S = np.trapz(np.sin(x**2), dx=dx)
# Parametric plot
t = np.linspace(0, 2*np.pi, 1000)
S = np.sin(t)
C = np.cos(t)

# Scale the values
S *= L/2
C *= L/2

plt.plot(C, S, label='Parametric Plot')
plt.xlabel('C(z)')
plt.ylabel('S(z)')
plt.title('Cornu Spiral')
plt.legend()
plt.show()
```

```python
#4a
import numpy as np
import matplotlib.pyplot as plt

# function definition
f = lambda x: x**3 - 100*np.cos(x)

# finding the root using Newton's Method
x = 1 # initial guess
for i in range(100): # iterations
    x = x - f(x)/(3*x**2 - 100*np.sin(x))
    if abs(f(x)) < 1e-6:
        break

print("Root found by Newton's Method: ", x)

# plotting the function
x_values = np.linspace(-1, 4, 1000)
plt.plot(x_values, f(x_values))
plt.title('Plot of x^3 - 100*cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
#4b
# function definition
f = lambda x: np.exp(x) * np.sin(x) - x**22 + 5

# finding the maximum value using the Bisection Method
x1 = -1
x2 = 3
while abs(x2 - x1) > 1e-6:
    x_mid = (x1 + x2) / 2
    if f(x_mid) > f((x1 + x2) / 2):
        x1 = x_mid
    else:
        x2 = x_mid

print("Maximum value of f(x) found by Bisection Method: ", f(x1))
print("x at maximum value: ", x1)

# First derivative of f(x)
f_prime = lambda x: np.exp(x) * np.sin(x) + np.exp(x) * np.cos(x) - 2*x*np.exp(x) * np.sin(x)

# Second derivative of f(x)
f_double_prime = lambda x: np.exp(x) * np.sin(x) + np.exp(x) * np.cos(x) - 2*x*np.exp(x) * np.cos(x) + 2*x*np.exp(x) * np.sin(x) - 2*x*np.exp(x) * np.sin(x)

# Find the maximum value of f(x) using the second derivative test
x_values = np.linspace(-1, 3, 1000)
extrema_values = []
for i in range(len(x_values) - 1):
    if f_double_prime(x_values[i]) < 0:
        extrema_values.append((x_values[i], x_values[i+1]))

maximum_values = []
for extrema in extrema_values:
    if f(extrema[0]) > f(extrema[1]):
        maximum_values.append((extrema[0], f(extrema[0])))
    else:
        maximum_values.append((extrema[1], f(extrema[1])))

print("Maximum values of f(x) found by the second derivative test: ", maximum_values)
```

```python
#6
import math

def gamma(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return (n - 1) * gamma(n - 1)

def main():
    n = 12
    gamma_value = gamma(n)
    sqrt_pi = math.sqrt(math.pi)
    print(f"{n}2 = {gamma_value}")
    print(f"  {sqrt_pi}")

    n = 72
    gamma_value = gamma(n)
    print(f"{n}2 = {gamma_value}")
    print(f"{n}2  {gamma_value/math.pow(10, 25)}")
main()
```
