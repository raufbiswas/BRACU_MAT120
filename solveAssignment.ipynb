```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, trapz
from scipy.misc import derivative
def f(x):
    return 2*x*(x-3)*(np.sin(x)-5)

def h(t):
    return t**2
def dfdx(x):
    return derivative(f, x)
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
t = np.linspace(-4, 4, 1000)

plt.figure(figsize=(12, 6))

# Plot f(x)
plt.subplot(121)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, dfdx(x), label="f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f(x) and its derivative')
plt.legend()

# Plot h(t)
plt.subplot(122)
plt.plot(t, h(t), label='h(t)')
plt.plot(t, derivative(h, t), label="h'(t)")
plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of h(t) and its derivative')
plt.legend()

plt.show()
```

```python
# Trapezoidal rule for f(x)
result_trapezoidal_rule_f = trapz(f(x), x)

# quad for f(x)
result_quad_f = quad(f, -2*np.pi, 2*np.pi)[0]

# Trapezoidal rule for h(t)
result_trapezoidal_rule_h = trapz(h(t), t)

# quad for h(t)
result_quad_h = quad(h, -4, 4)[0]
print(f"Result from Trapezoidal rule for f(x): {result_trapezoidal_rule_f}")
print(f"Result from quad for f(x): {result_quad_f}")
print(f"Result from Trapezoidal rule for h(t): {result_trapezoidal_rule_h}")
print(f"Result from quad for h(t): {result_quad_h}")
```

```python
import numpy as np

def f(z):
    return np.sin(2*z) - 3

def simpsons_rule(f, a, b, N):
    h = (b - a) / N
    xi = a
    sum = f(a) + f(b)
    
    for i in range(1, N):
        xi += h
        if i % 2 == 0:
            sum += 2 * f(xi)
        else:
            sum += 4 * f(xi)
    
    return (h / 3) * sum

# Define the required parameters
a = 0
b = 2.5
N = 20

# Compute the integral
integral = simpsons_rule(f, a, b, N)

print(f"The integral of f(z) from {a} to {b} using Simpson's Rule is: {integral}")
```
