```python
import sympy as sp
from sympy import *
sp.init_printing()
x,y,z = sp.symbols('x y z')
```

```python
#task1
f = sp.Function('f')(x)
expr1 = sp.Eq(x*sp.diff(f,x,2)+sp.diff(f,x,1), x**3)
expr1
```

```python
solve1 = sp.dsolve(expr1)
display(solve1)
```

```python
solve1 = sp.dsolve(expr1, ics = {f.subs(x,1): '0', sp.diff(f,x,1).subs(x,2): '1'})
solve1
```

```python
#task2
y = sp.Function('y')(x)
expr2 = sp.Eq(sp.diff(y,x,1), (-y**2/x**2))
expr2
```

```python
solve2 = sp.dsolve(expr2)
display(solve2)
```

```python
solve2 = sp.dsolve(expr2, ics= {y.subs(x,1): '1/2'})
solve2
```
